```python
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Load dataset
file_path = "laptop_prices.csv"
df = pd.read_csv(file_path)

# Selecting necessary features
features = ["Company", "OS", "CPU_company", "PrimaryStorageType", "SecondaryStorageType", "GPU_company",
            "Inches", "Ram", "Weight", "ScreenW", "ScreenH", "CPU_freq", "PrimaryStorage", "SecondaryStorage"]
target = "Price_euros"

df = df[features + [target]].dropna()

# Encoding categorical features
label_encoders = {}
for col in ["Company", "OS", "CPU_company", "PrimaryStorageType", "SecondaryStorageType", "GPU_company"]:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Splitting dataset
X = df[features]
y = df[target]

# Training model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# Save model and encoders
joblib.dump(model, "laptop_price_model.joblib")
joblib.dump(label_encoders, "label_encoders.joblib")

print("Model training complete and saved.")
```

```python
import pandas as pd
import joblib

# Load model and encoders
model = joblib.load("laptop_price_model.joblib")
label_encoders = joblib.load("label_encoders.joblib")

# Function to get valid user input
def get_valid_input(prompt, valid_options=None, value_range=None, dtype=str):
    while True:
        if valid_options:
            print(f"Available options: {', '.join(valid_options)}")  # Show available choices
        
        user_input = input(prompt).strip()  # Remove extra spaces
        
        if dtype == str:
            # Match user input with available options (case-insensitive)
            if valid_options:
                matched_option = next((opt for opt in valid_options if opt.lower() == user_input.lower()), None)
                if matched_option:
                    user_input = matched_option
                else:
                    print(f"Invalid choice. Choose from: {', '.join(valid_options)}")
                    continue
        
        try:
            if dtype in [int, float]:
                user_input = dtype(user_input)
                if value_range and not (value_range[0] <= user_input <= value_range[1]):
                    print(f"Please enter a value between {value_range[0]} and {value_range[1]}.")
                    continue
            return user_input
        except ValueError:
            print(f"Please enter a valid {dtype.__name__}.")

# Feature details (categorical values and numeric ranges)
feature_options = {
    "Company": list(label_encoders["Company"].classes_),
    "OS": list(label_encoders["OS"].classes_),
    "CPU_company": list(label_encoders["CPU_company"].classes_),
    "PrimaryStorageType": list(label_encoders["PrimaryStorageType"].classes_),
    "SecondaryStorageType": list(label_encoders["SecondaryStorageType"].classes_),
    "GPU_company": list(label_encoders["GPU_company"].classes_)
}

numeric_ranges = {
    "Inches": (10, 20),
    "Ram": (2, 64),
    "Weight": (0.5, 5),
    "ScreenW": (800, 3840),
    "ScreenH": (600, 2160),
    "CPU_freq": (1.0, 5.0),
    "PrimaryStorage": (128, 8192),
    "SecondaryStorage": (0, 8192)
}

# Get user input
user_data = {}
for feature, options in feature_options.items():
    user_data[feature] = get_valid_input(f"Enter {feature}: ", valid_options=options, dtype=str)

for feature, (min_val, max_val) in numeric_ranges.items():
    user_data[feature] = get_valid_input(f"Enter {feature} ({min_val}-{max_val}): ", value_range=(min_val, max_val), dtype=float if feature == "Weight" else int)

# Convert user input to DataFrame
input_df = pd.DataFrame([user_data])

# Encode categorical variables
for col in feature_options.keys():
    input_df[col] = label_encoders[col].transform(input_df[col])

# Predict price
predicted_price = model.predict(input_df)[0]
print(f"Predicted Laptop Price: â‚¬{predicted_price:.2f}")
```

```python
from PIL import Image, ImageDraw, ImageFont

def generate_price_image(user_input_text, predicted_price, output_filename="laptop_price.png"):
    # Create an image
    img = Image.new("RGB", (800, 600), color="white")
    draw = ImageDraw.Draw(img)

    # Load font (default system font)
    try:
        font = ImageFont.truetype("arial.ttf", 20)
    except:
        font = ImageFont.load_default()

    # Prepare text
    text = f"{user_input_text}\nPredicted Laptop Price: â‚¬{predicted_price:.2f}"
    
    # Draw text
    draw.multiline_text((20, 20), text, fill="black", font=font)

    # Save the image
    img.save(output_filename)
    print(f"Image saved as {output_filename}")

# Example usage
user_input_text = """
Available options: Acer, Apple, Asus, Chuwi, Dell, Fujitsu, Google, HP, Huawei, LG, Lenovo, MSI, Mediacom, Microsoft, Razer, Samsung, Toshiba, Vero, Xiaomi
Enter Company: Acer
Enter OS: Windows 10
Enter CPU_company: AMD
Enter PrimaryStorageType: SSD
Enter SecondaryStorageType: HDD
Enter GPU_company: Intel
Enter Inches: 15
Enter Ram: 16
Enter Weight: 1.3
Enter ScreenW: 1200
Enter ScreenH: 1200
Enter CPU_freq: 4
Enter PrimaryStorage: 256
Enter SecondaryStorage: 512
"""

predicted_price = 2191.04
generate_price_image(user_input_text, predicted_price)
```
